```python
import re
from copy import deepcopy
from pathlib import Path
from lang_id import Name2Abbr, L2L





data_dir = Path("/data/duhu/dbase/FangYan")

projs = []
langs = set()
proj2name = {}
for folder in data_dir.glob("*"):
    if not folder.is_dir():
        continue
    proj = folder.stem
    projs.append(proj)
    if re.findall("[a-zA-Z]", proj.split("-")[-1]):
        proj_ = "-".join(proj.split("-")[:-1])
    else:
        proj_ = proj
    lang = Name2Abbr[L2L[proj_]]
    proj2name[proj] = lang
    
    langs.add(f"<{lang.upper()}>")
    is_read = False
    table_dir = folder / "TABLE"
    if table_dir.is_dir():
        x = f"""  {folder.stem}:
    type: read
    conf:
      data_dir: /data/duhu/dbase/FangYan
      kformat_dir: data
      mode: train
      test_rate: 0.001
      item_index: 1
      lang: {lang}"""
    else:
        x = f"""  {folder.stem}:
    type: diag
    conf:
      data_dir: /data/duhu/dbase/FangYan
      kformat_dir: data
      mode: train
      test_rate: 0.001
      item_index: 0
      lang: {lang}"""
    print(x+"\n")
```

```python
from collections import defaultdict

proj2lang = {
    'King-ASR-426': 'ZangYu',
    'King-ASR-384-14-Liaoning': 'LiaoNing',
    'King-ASR-406': 'SiChuan',
    'King-ASR-399': 'Yue_HK',
    'King-ASR-384-1-Fujian': 'FuJian',
    'King-ASR-345': 'WeiYu',
    'King-ASR-853-1': 'MinNan',
    'King-ASR-384-13-Henan': 'HeNan',
    'King-ASR-384-19-Wenzhou': 'WenZhou',
    'King-ASR-086': 'Yue_HK',
    'King-ASR-407': 'ShangHai',
    'King-ASR-384-3-Hunan': 'HuNan',
    'King-ASR-742': 'ZangYu',
    'King-ASR-384-23-Ningxia': 'NingXia',
    'King-ASR-384-21-Liaoning': 'LiaoNing',
    'King-ASR-384-15-Shaanxi': 'ShanXi',
    'King-ASR-083': 'Yue_HK',
    'King-ASR-384-11-Anhui': 'AnHui',
    'King-ASR-443': 'Yue_HK',
    'King-ASR-881': 'SiChuan',
    'King-ASR-423': 'Yue_GZ',
    'King-ASR-422': 'ZangYu',
    'King-ASR-241': 'Yue_GZ',
    'King-ASR-384-10-Tianjin': 'TianJin',
    'King-ASR-446': 'Yue_HK',
    'King-ASR-384-2': 'GuangDong',
    'King-ASR-882': 'SiChuan',
    'King-ASR-854': 'MinNan',
    'King-ASR-384-22-Wu': 'WuYu',
    'King-ASR-424': 'WeiYu',
    'King-ASR-070': 'Yue_HK',
    'King-ASR-854-1': 'MinNan',
    'King-ASR-879': 'Yue_GZ',
    'King-ASR-384-12-Shandong': 'ShanDong',
    'King-ASR-420': 'MinNan',
    'King-ASR-758': 'Yue_HK',
    'King-ASR-384-18-Gansu': 'GanSu',
    'King-ASR-427': 'MinNan',
    'King-ASR-384-20-Hebei': 'HeBei',
    'King-ASR-384-6-WU': 'WuYu',
    'King-ASR-384-4-Jiangxi': 'JiangXi',
    'King-ASR-791': 'ShangHai',
    'King-ASR-384-7-Yunnan': 'YunNan',
    'King-ASR-384-17-Hubei': 'HuBei',
    'King-ASR-384-16-Shanxi': 'ShanXi',
    'King-ASR-384-8-Guizhou': 'GuiZhou',
    'King-ASR-880': 'Yue_GZ',
    'King-ASR-883': 'SiChuan',
    'King-ASR-384-9-Wu': 'WuYu',
    'King-ASR-669': 'Yue_GZ',
    'King-ASR-668': 'Yue_GZ',
    'King-ASR-449': 'Yue_HK',
    'King-ASR-421': 'WeiYu',
    'King-ASR-853': 'MinNan'
}

lang2projs = defaultdict(list)
for proj, lang in proj2lang.items():
    lang2projs[lang].append(proj)
proj2durs = defaultdict(list)
for index in range(0, len(cons), 3):
    proj, test, train = cons[index].strip(), cons[index+1].strip(), cons[index+2].strip()
    try:
        test_dur = float(test.split(" ")[-1]) / 3600
        train_dur = float(train.split(" ")[-1]) / 3600
        proj2durs[proj] = [train_dur, test_dur]
    except Exception:
        print(proj)
```

```python
f = open("details.txt", 'w')
lang2dur = defaultdict(lambda : [0, 0])
for lang, projs in sorted(lang2projs.items(), key=lambda x:x[0]):
    for proj in projs:
        try:
            train, test = proj2durs[proj]
            lang2dur[lang][0] += train
            lang2dur[lang][1] += test
        except Exception:
            train, test = "/", "/"
        f.write(f"{lang}\t{proj}\t{train}\t{test}\n")
f.close()
f1 = open('merge.txt', 'w')

for lang, dur in lang2dur.items():
    print(lang, dur[0], dur[1])
    f1.write(f"{lang}\ts1\t{round(dur[0], 3)}\t{round(dur[1], 3)}\n")
f1.close()
```
